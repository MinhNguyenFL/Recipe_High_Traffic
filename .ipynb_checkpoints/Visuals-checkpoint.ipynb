```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
df = pd.read_csv('recipe_site_traffic_2212.csv')

df.high_traffic = df.high_traffic.fillna('Not High')
df['category'] = df['category'].replace('Chicken Breast', 'Chicken')
df['servings'] = df['servings'].str.replace(' as a snack', '').astype(int)

def fill_missing(cat):
    df[cat] = df.groupby('category')[cat].transform(lambda x: x.fillna(x.median()))

for x in ['calories', 'carbohydrate', 'sugar', 'protein']:
    fill_missing(x)
df.isna().sum()


# encode high_traffic , High = 1
df['high_traffic'] = df['high_traffic'].map({'High': 1, 'Not High': 0})

# One hot encode
from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder object
encoder = OneHotEncoder()
# Fit and transform the "category" column
encoded_categories = encoder.fit_transform(df[['category']])
# Create a DataFrame with the encoded categories
encoded_df = pd.DataFrame(encoded_categories.toarray(), columns=encoder.categories_[0])
# Concatenate the original DataFrame with the encoded DataFrame
df_encoded = pd.concat([df, encoded_df], axis=1)
# Drop columns
df_encoded.drop(['category','Vegetable','recipe'], axis=1, inplace=True)
# Print the encoded data
df_encoded.head()
```

```python
features = df_encoded.drop('high_traffic', axis = 1)
target = df_encoded['high_traffic']

# Import the function for splitting dataset into train and test
from sklearn.model_selection import train_test_split

# Use that function to create the splits both for target and for features
target_train, target_test, features_train, features_test = train_test_split(target,features,test_size=0.25,random_state=42)
```
